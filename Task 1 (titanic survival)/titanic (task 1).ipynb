# Prediction of survival rate on titanic dataset 

## Column description

- **PassengerId:** This is a unique number provided to every guest. It functions similarly to an ID number in tracking everyone.

- **Survived:** This tells us whether the passenger survived the Titanic accident. A 0 signifies they did not survive, while a 1 means they did.

- **Pclass:** This represents the class of the passenger's ticket. There are three classes: the first, second, and third class. First class was the most luxury, while third class was the least expensive.
- **Name:** This is the passenger's full name.
- **Sex:** This identifies the passenger's gender, whether male or female.
- **Age:** This is the passenger's age in years. Some ages may be missing because they were not recorded.
- **SibSp:** This refers to siblings/spouses living abroad. It indicates how many siblings or spouses the passenger had aboard the Titanic.
- **Parch:** This is for parents/children on board. It indicates how many parents or children accompanied the guest on the voyage.
- **Ticket:** This is the ticket number used by the traveler to board the Titanic.
- **Fare:** This is the amount of money the passenger paid for the ticket.
- **Cabin:** This is the stateroom number where the passenger stayed on the ship. Many travelers have no documented cabin number.
- **Embarked:** This is the port where passengers boarded the Titanic. The ports are: C = Cherbourg. Q = Queenstown. S denotes Southampton.

```python
#Dropping the 'Cabin' column from the dataframe
df.drop(['Cabin'], axis=1, inplace=True)
```

```python
#Dropping the 'Cabin' column from the dataframe
df.drop(['PassengerId'], axis=1, inplace=True)
```

```python
# converting categorical Columns
df.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
df['Embarked'] = df['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
```

```python
#Copying data so that original dataset is not affected
data = df[['Survived', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Age', 'Fare']].copy(deep= True)
```

```python
# importing train test model and loading independent variables to x and dependent variables to y
from sklearn.model_selection import train_test_split
x = data[['Pclass', 'Sex', 'Parch', 'Age','Fare']]
y = data['Survived']
```

```python
# assigning variables in train test split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
# importing classification models
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# KNeighbhors algorithm where we have kept value of k in a loop to get an optimum value with least misclassified data
missclassified=[]
for i in range(1,20):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    kpred = knn.predict(x_test)
    missclassified.append((y_test != kpred).sum())
print(missclassified)
```

```python
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(x_train,y_train)
kpred = knn.predict(x_test)
kcm = confusion_matrix(y_test,kpred)
kacc = accuracy_score(y_test,kpred)
print(f'The confusion matrix is:\n{kcm}\nThe accuracy is: {round((kacc*100), 2)}%')
```

```python
# Logistic Regression algorithm
lr = LogisticRegression()
lr.fit(x_train,y_train)
lpred = lr.predict(x_test)
lcm = confusion_matrix(y_test,lpred)
lacc = accuracy_score(y_test,lpred)
print(f'The confusion matrix is:\n{lcm}\nThe accuracy is: {lacc*100}%')
```

```python
# Decision Tree algorithm
dtc = DecisionTreeClassifier()
dtc.fit(x_train,y_train)
dpred = dtc.predict(x_test)
dcm = confusion_matrix(y_test,dpred)
dacc = accuracy_score(y_test,dpred)
print(f'The confusion matrix is:\n{dcm}\nThe accuracy is: {dacc*100}%')
```

```python
# Random Forest algorithm
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
rpred = rfc.predict(x_test)
rcm = confusion_matrix(y_test,rpred)
racc = accuracy_score(y_test,rpred)
print(f'The confusion matrix is:\n{rcm}\nThe accuracy is: {racc*100}%')
```

```python
# Naives Bayes - Gaussian theorem algorithm
nb = GaussianNB()
nb.fit(x_train,y_train)
npred = nb.predict(x_test)
ncm = confusion_matrix(y_test,npred)
nacc = accuracy_score(y_test,npred)
print(f'The confusion matrix is:\n{ncm}\nThe accuracy is: {nacc*100}%')
```

```python
# Scalar Vector Machine algorithm
svm = SVC(kernel='linear',random_state=12)
svm.fit(x_train,y_train)
spred = svm.predict(x_test)
scm = confusion_matrix(y_test,spred)
sacc = accuracy_score(y_test,spred)
print(f'The confusion matrix is:\n{scm}\nThe accuracy is: {sacc*100}%')
```
