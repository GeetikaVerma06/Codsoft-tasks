# Credit card fraud detection

This is a column description for the dataset.

1. **Time**: The number of seconds since the first transaction in the dataset. It indicates when the transaction occurred in comparison to others.

2. **V1–V28**: These are the principle components obtained by PCA (principle Component Analysis) to safeguard the confidentiality of the data. Each of these variables combines the dataset's original attributes.

3. **Amount**: The whole monetary worth of the transaction.

4. **Class**: The variable that determines whether the transaction is fraudulent (1) or not (0).

# Train and test splitting

```python
# copying data such that original is not affected and selecting columns that are relevent
data = df.copy(deep= True)
```

```python
# dropping duplicate records
data.drop_duplicates(keep='first', inplace = True)
```

```python
# since amount column has high standard deviation we need to preprocess it using Standard Scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data['Amount'] = scaler.fit_transform(data['Amount'].values.reshape(-1,1))
```

```python
# dropping time column as there is no relevance of transaction time with fraud detection
data.drop(columns= 'Time', inplace= True)
```

```python
# splitting data into dependent and independent
X = data.drop(columns='Class',axis = 1)
Y = data['Class']
```

```python
# import libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix)
```

```python
X_train,X_test, Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 42)
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model on the training data
model.fit(X_train,Y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)
```

```python
print("Confusion Matrix:\n", confusion_matrix(Y_test, y_pred))
print("\nClassification Report:\n", classification_report(Y_test, y_pred))
print("\nAccuracy Score:", accuracy_score(Y_test, y_pred))
```
